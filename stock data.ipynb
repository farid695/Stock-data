```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Extract Tesla stock data
tesla_data = yf.download('TSLA', start='2019-01-01', end='2024-01-01')

# Step 2: Reset index to make Date a column
tesla_data.reset_index(inplace=True)

# Step 3: Define the make_graph function
def make_graph(data):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', color='blue')
    plt.title('Tesla Stock Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# Step 4: Call the make_graph function
make_graph(tesla_data)
```
